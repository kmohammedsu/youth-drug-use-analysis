```python
import pandas as pd
import matplotlib.pyplot as plt

# Load processed data
df = pd.read_csv("../data/processed_data.csv")

# Display dataset overview
print("Processed Dataset Overview:")
print(df.info())

# Binary classification results
binary_cm_img = "../results/binary_confusion_matrix.png"
binary_tree_img = "../results/binary_decision_tree.png"

# Display confusion matrix
plt.figure(figsize=(10, 6))
img = plt.imread(binary_cm_img)
plt.imshow(img)
plt.axis("off")
plt.title("Binary Classification: Confusion Matrix")
plt.show()

# Display decision tree visualization
plt.figure(figsize=(10, 6))
img = plt.imread(binary_tree_img)
plt.imshow(img)
plt.axis("off")
plt.title("Binary Classification: Decision Tree")
plt.show()
```

```python
# Multi-class classification results
multi_cm_img = "../results/multi_class_confusion_matrix.png"
multi_tree_img = "../results/multi_class_decision_tree.png"
multi_feature_img = "../results/multi_class_feature_importance.png"

# Display confusion matrix
plt.figure(figsize=(10, 6))
img = plt.imread(multi_cm_img)
plt.imshow(img)
plt.axis("off")
plt.title("Multi-Class Classification: Confusion Matrix")
plt.show()

# Display decision tree visualization
plt.figure(figsize=(10, 6))
img = plt.imread(multi_tree_img)
plt.imshow(img)
plt.axis("off")
plt.title("Multi-Class Classification: Decision Tree")
plt.show()

# Display feature importance chart
plt.figure(figsize=(10, 6))
img = plt.imread(multi_feature_img)
plt.imshow(img)
plt.axis("off")
plt.title("Multi-Class Feature Importance")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# File paths for saved visualizations
regression_tree_img = "../results/regression_decision_tree.png"
regression_plot_img = "../results/regression_predicted_vs_actual.png"

# Display decision tree visualization
plt.figure(figsize=(25, 15))
img = mpimg.imread(regression_tree_img)
plt.imshow(img)
plt.axis("off")
plt.title("Regression: Decision Tree for Age of First Cigarette Use")
plt.show()

# Display actual vs. predicted scatter plot
plt.figure(figsize=(10, 6))
img = mpimg.imread(regression_plot_img)
plt.imshow(img)
plt.axis("off")
plt.title("Regression: Actual vs Predicted Age of First Cigarette Use")
plt.show()
```

### Summary of Modeling Results

#### Binary Classification (Cigarette Use):
- **Accuracy**: 65%
- **Precision for Class 1 (Ever Used)**: 2%
- **Recall for Class 1**: 52%
- **Key Predictors**:
  - Parental Help (`PARHLPHW`): 54.8%
  - School Safety (`SCHFELT`): 31.1%
  - Income (`INCOME`): 14.1%

#### Multi-Class Classification (Marijuana Frequency):
- **Accuracy**: 75%
- **Precision for "Seldom" Class**: 17%
- **Recall for "Seldom" Class**: 53%
- **Key Predictors**:
  - Peer Influence (`PEER_INFLUENCE`): 43.9%
  - Health Status (`HEALTH2`): 15.9%
  - Risk Behaviors (`YOGRPFT2`): 15.9%

#### Regression (Age of First Cigarette Use):
- **Mean Squared Error (MSE)**: 5.8579
- **Root Mean Squared Error (RMSE)**: 2.4203
- **R² Score**: 0.0328
- **Key Predictors**:
  - Income Level (`INCOME`)
  - Peer Influence (`PEER_INFLUENCE`)
  - Standardized Alcohol Attitudes (`STNDALC`)
  - Risk-Taking Behavior (`RISK_TAKING`)
